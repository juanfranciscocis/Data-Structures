```python
def sumOfArr(arr):
    sums = []
    sum = 0
    for i in arr:
        if(str(i) == ""):
            print(sum)
            sums.append(sum)
            sum = 0
            continue


        sum+= int(i)

    sums.sort()
    longitud = len(sums)

    top3 = []

    for j in range(3):
        top3.append(sums[longitud-1])
        longitud-=1


    sumTop3 = 0

    for k in top3:
        sumTop3+=k








    return sumTop3




```
